```python
from bs4 import BeautifulSoup

html = """
<html>
    <body>
        <h1>Welcome to OpenAI</h1>
        <p class="lead">Creating artificial general intelligence for the benefit of all.</p>
    </body>
</html>
"""

soup = BeautifulSoup(html, "html.parser")
header_text = soup.h1.text
paragraph_text = soup.find("p", class_="lead").text

print("Header: ", header_text)
print("Paragraph: ", paragraph_text)
```

```python
link = soup.find("a")

if link:
    href = link['href']
    print("Link: ", href)
else:
    print("No anchor tag found in the HTML.")
```

```python
div = soup.find('div', class_='content')
if div:
  paragraphs = div.find_all('p') # div 내의 모든 p 요소를 찾습니다
  for p in paragraphs:
    print(p.text)

else:
  print("No <div> tag with class 'content' found.")
```

```python
table = soup.find('table')
if table:
  for row in table.find_all('tr'):
    cells = row.find_all('td')
    for cell in cells:
      print(cell.text)
    # 'cells'는 해당 tr 요소 내의 모든 td 요소를 포함합니다
else:
  print("No table tag found")

# 목록의 경우
list = soup.find('ul')
if list:
  for item in list.find_all('li'):
    print(item.text) # 각 목록 항목의 텍스트를 출력합니다
else:
  print("No ui tag found")
```

```python
import requests

# API 엔드포인트 정의
url = "https://api.weather.com/v1/current"

# 쿼리 매개변수 정의
params = {
    "location": "San Francisco",
    "API_KEY":
}

# GET 요청 보내기
response = requests.get(url, params=params)

# 상태 코드 출력 (모든 것이 정상이라면 200이어야 함)
print(response.status_code)

if response.status_code == 200:
  # API에서 반환된 데이터 출력
  print(response.json())
else:
  print("APi False",response.status_code)

try:
    data = response.json()
except ValueError as e:
    print("JSON 디코딩 오류:", e)
    print("응답 내용:", response.text)
```

```python
import sqlite3

conn = sqlite3.connect('data.db')

cur = conn.cursor()

cur.execute("""CREATE TABLE IF NOT EXISTS user_data (
    name TEXT,
    age INTEGER,
    country TEXT
    )""")

data = [("John Doe", 25, "미국"), ("Jane Smith", 28, "캐나다")]
cur.executemany("""INSERT INTO user_data VALUES (?, ?, ?)""",data)

conn.commit()
conn.close()
```
