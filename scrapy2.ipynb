```python
# 키보드 엔터

options = webdriver.ChromeOptions()
options.add_argument('--headless')  #헤드리스
options.add_argument('--no-sandbox')  # 샌드박스 비활성화
options.add_argument('--disable-dev-shm-usage')  # /dev/shm 비활성화
options.add_argument('--disable-dev-shm-usage')  # 사용자 데이터 디렉토리 설정

driver = webdriver.Chrome(options=options)

try:
    driver.get("https://www.naver.com")

    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "#query")))

    search_box = driver.find_element(By.CSS_SELECTOR, "#query")
    search_box.send_keys("코로나")

    search_button = driver.find_elemeiotytnt(By.CSS_SELECTOR, "#search_btn")
    search_button.click()

finally:
    driver.quit()
```
